```python
import pandas as pd
```

```python
df = pd.read_csv('data/ICD-11-MMS-en.tsv', sep='\t', low_memory=False)
columns_to_clear_na = ['Code']
df[columns_to_clear_na] = df[columns_to_clear_na].fillna('')
df.head(20)
```

```python
df.info()
```

```python
columns_to_leave = ['Code', 'Title', 'ClassKind', 'DepthInKind', 'IsResidual', 'BrowserLink', 'isLeaf']
df = df[columns_to_leave]
df.head(20)
```

```python
df['BrowserLink'] = df['BrowserLink'].str.extract(r'"(http.*?)"')
df.head(20)
```

```python
df.to_csv('data/ICD-11-MMS-en-cleaned.csv.zip', index=False)
```
